```python
numeric_cols = df.select_dtypes(include=['float64', 'int64'])
outliers = {}

# IQR method to detect outliers
for col in numeric_cols.columns:
    Q1 = numeric_cols[col].quantile(0.25)
    Q3 = numeric_cols[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outlier_count = numeric_cols[(numeric_cols[col] < lower) | (numeric_cols[col] > upper)].shape[0]
    if outlier_count > 0:
        outliers[col] = outlier_count

# Print the outlier counts
for col, count in outliers.items():
    print(f"{col}: {count} outliers")
plt.figure(figsize=(16, 8))
numeric_cols.boxplot(rot=90)
plt.title("Outlier Detection")
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
df = df.dropna(subset = ['States'])                  # without subset, dropna will remove all rows where any column entry is missing
df
```

```python
# Strip column names of spaces just in case
df.columns = df.columns.str.strip()

# Fill numeric columns with 0
for col in ['confirmed', 'deaths', 'recovered', 'vaccines', 'people_vaccinated', 'people_fully_vaccinated']:
    if col in df.columns:
        df.loc[:, col] = df[col].fillna(0)

# Fill 'tests' with median
if 'tests' in df.columns:
    df.loc[:, 'tests'] = df['tests'].fillna(df['tests'].median())

# Fill policy-related columns with mode
policy_cols = [
    'school_closing', 'workplace_closing', 'cancel_events',
    'gatherings_restrictions', 'transport_closing', 'stay_home_restrictions',
    'information_campaigns', 'testing_policy', 'vaccination_policy',
    'elderly_people_protection'
]

for col in policy_cols:
    if col in df.columns:
        df.loc[:, col] = df[col].fillna(df[col].mode()[0])
```

### 1. Trend analysis of COVID-19 cases

```python
# Group data by date and sum cases
trend_data = df.groupby('date')[['confirmed', 'deaths', 'recovered']].sum()
# Plot the area chart
plt.figure(figsize=(10,5))
sns.set_style("darkgrid")

plt.fill_between(trend_data.index, trend_data['confirmed'], color='blue', alpha=0.3, label="Confirmed Cases")
plt.fill_between(trend_data.index, trend_data['deaths'], color='red', alpha=0.5, label="Deaths")
plt.fill_between(trend_data.index, trend_data['recovered'], color='green', alpha=0.3, label="Recovered Cases")

# Aesthetic improvements
plt.xlabel("Date", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.title("COVID-19 Cases Trend in India (Area Chart)", fontsize=14, fontweight="bold")
plt.xticks(rotation=45)
plt.legend(fontsize=10)
```

```python
df = df.copy()
df['year'] = df['date'].dt.year

yearly = df.groupby('year')[['confirmed', 'deaths', 'recovered']].sum()
# Step 6: Filter for 2020 and 2021

yearly = yearly.loc[[2020, 2021]]
# Step 7: Plot
fig, ax = plt.subplots(figsize=(8, 6))
yearly.plot(kind='bar', stacked=True, colormap='viridis', ax=ax)
plt.title("COVID-19 Cases Comparison (2020 vs 2021)", fontsize=14, fontweight="bold")
plt.xlabel("Year", fontsize=12)
plt.ylabel("Total Cases", fontsize=12)
plt.legend(title="Category")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### 2. Comparing COVID -19 cases month wise in different states and union terrirtory of India 

```python
df = df.copy()
df['year_month'] = df['date'].dt.to_period('M')
df['year_month_str'] = df['date'].dt.strftime('%b %Y')

# Define States and UTs
states = [
    'Andhra Pradesh', 'Arunachal Pradesh', 'Assam', 'Bihar', 'Chhattisgarh', 'Goa',
    'Gujarat', 'Haryana', 'Himachal Pradesh', 'Jharkhand', 'Karnataka', 'Kerala',
    'Madhya Pradesh', 'Maharashtra', 'Manipur', 'Meghalaya', 'Mizoram', 'Nagaland',
    'Orissa', 'Punjab', 'Rajasthan', 'Sikkim', 'Tamil Nadu', 'Telagana',
    'Tripura', 'Uttar Pradesh', 'Uttaranchal', 'West Bengal'
]

uts = [
    'Andaman and Nicobar Islands', 'Chandigarh',
    'Dadra and Nagar Haveli and Daman and Diu', 'Delhi', 'Jammu and Kashmir',
    'Ladakh', 'Lakshadweep', 'Puducherry'
]
df['States'] = df['States'].str.strip()
df['region_type'] = df['States'].apply(lambda x: 'State' if x in states else 'UT' if x in uts else 'Unknown')
monthly_data = df.groupby(['region_type', 'year_month_str', 'States'])['confirmed'].sum().reset_index()
state_data = monthly_data[monthly_data['region_type'] == 'State']
ut_data = monthly_data[monthly_data['region_type'] == 'UT']

state_table = state_data.pivot(index='year_month_str', columns='States', values='confirmed').fillna(0)
ut_table = ut_data.pivot(index='year_month_str', columns='States', values='confirmed').fillna(0)

for ut in uts:
    if ut not in ut_table.columns:
        ut_table[ut] = 0

state_table = state_table[sorted(state_table.columns)]
ut_table = ut_table[sorted(ut_table.columns)]

state_table.index = pd.to_datetime(state_table.index, format='%b %Y')
state_table.sort_index(inplace=True)
state_table.index = state_table.index.strftime('%b %Y')

ut_table.index = pd.to_datetime(ut_table.index, format='%b %Y')
ut_table.sort_index(inplace=True)
ut_table.index = ut_table.index.strftime('%b %Y')

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(16, 14), sharex=True)

state_table.plot(kind='bar', stacked=True, ax=ax1, colormap='plasma')
ax1.set_title("Monthly COVID-19 Confirmed Cases in Indian States", fontsize=20, fontweight='bold')
ax1.set_ylabel("Confirmed Cases",fontsize = 15)

ut_table.plot(kind='bar', stacked=True, ax=ax2, colormap='viridis')
ax2.set_title("Monthly COVID-19 Confirmed Cases in Union Territories", fontsize=20, fontweight='bold')
ax2.set_ylabel("Confirmed Cases", fontsize = 15)
plt.xticks(rotation=45)
plt.xlabel("Month",fontsize = 15)
plt.tight_layout()
```

```python
region_totals = df.groupby(['region_type', 'States'])['confirmed'].sum().reset_index()
top_states = region_totals[region_totals['region_type'] == 'State'].nlargest(10, 'confirmed')
top_uts = region_totals[region_totals['region_type'] == 'UT'].nlargest(5, 'confirmed')
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))

ax1.bar(top_states['States'], top_states['confirmed'], color='orange')
ax1.set_title("Top 10 States with Highest COVID-19 Confirmed Cases", fontsize=14, fontweight='bold')
ax1.set_ylabel("Confirmed Cases")
ax1.set_xticks(range(len(top_states)))
ax1.set_xticklabels(top_states['States'], rotation=45, ha='right')

ax2.bar(top_uts['States'], top_uts['confirmed'], color='teal')
ax2.set_title("Top 5 Union Territories with Highest COVID-19 Confirmed Cases", fontsize=14, fontweight='bold')
ax2.set_ylabel("Confirmed Cases")
ax2.set_xticks(range(len(top_uts)))
ax2.set_xticklabels(top_uts['States'], rotation=45, ha='right')
plt.tight_layout()
```

### 3. Analysing number of deaths and recovery rate

```python
# Group by date
daily_data = df.groupby('date')[['confirmed', 'deaths', 'recovered']].sum().reset_index()

# Calculate recovery rate
daily_data['recovery_rate'] = (daily_data['recovered'] / daily_data['confirmed']) * 100
daily_data['recovery_rate'].fillna(0)

# Set up plot
fig, ax1 = plt.subplots(figsize=(10, 5))

# Plot deaths
color1 = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Deaths', color=color1)
ax1.plot(daily_data['date'], daily_data['deaths'], color=color1, label='Deaths')
ax1.tick_params(axis='y', labelcolor=color1)

# Secondary axis for recovery rate
ax2 = ax1.twinx()
color2 = 'tab:green'
ax2.set_ylabel('Recovery Rate (%)', color=color2)
ax2.plot(daily_data['date'], daily_data['recovery_rate'], color=color2, label='Recovery Rate (%)')
ax2.tick_params(axis='y', labelcolor=color2)

# Add title and grid
plt.title('COVID-19: Deaths and Recovery Rate Over Time', fontsize=12, fontweight='bold')
ax1.grid(True, which='major', linestyle='--', alpha=0.5)

# Combine legends from both axes
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
fig.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

# Rotate x-axis labels
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
# Group by state/UT and calculate total confirmed, recovered, deaths and get population
statewise = df.groupby('States')[['confirmed', 'deaths', 'recovered', 'population']].sum().reset_index()

# Calculate death rate per 100k and recovery rate
statewise['death_rate_per_100k'] = (statewise['deaths'] / statewise['population']) * 100000
statewise['recovery_rate'] = (statewise['recovered'] / statewise['confirmed']) * 100

statewise_sorted = statewise.sort_values('death_rate_per_100k', ascending=False)
fig, ax = plt.subplots(figsize=(12, 10))

ax.barh(statewise_sorted['States'], statewise_sorted['death_rate_per_100k'], color='crimson', label='Death Rate per 100k')
ax2 = ax.twiny()
ax2.plot(statewise_sorted['recovery_rate'], statewise_sorted['States'], 'go', label='Recovery Rate (%)')
ax.set_xlabel('Death Rate per 100,000', fontsize=12)
ax2.set_xlabel('Recovery Rate (%)', fontsize=12)
ax.set_title('COVID-19 Death Rate vs Recovery Rate (Relative to Population)', fontsize=14, fontweight='bold')
ax.legend(loc='lower right')
ax2.legend(loc='upper right')
plt.tight_layout()
```

### 4. Analysing vaccinations across different states and union territory

```python
# Group by States and sum up vaccination data
vax = df.groupby('States')[['people_vaccinated', 'people_fully_vaccinated']].sum().reset_index()
vacc = vax.sort_values('people_vaccinated', ascending=False)
fig, ax = plt.subplots(figsize=(14,8))

bar_width = 0.4
x = range(len(vacc))

# Plot both bars side-by-side
ax.bar(x, vacc['people_vaccinated'], width=bar_width, label='People Vaccinated (1st Dose)', color='skyblue')
ax.bar([i + bar_width for i in x], vacc['people_fully_vaccinated'], width=bar_width, label='Fully Vaccinated', color='seagreen')

# Labels & Title
ax.set_xticks([i + bar_width/2 for i in x])
ax.set_xticklabels(vacc['States'], rotation=90)
ax.set_ylabel("Number of People")
ax.set_title("Vaccination Coverage Across States and Union Territory", fontsize=14, fontweight='bold')
ax.legend()
```

```python
# Graph between vaccinated people and remaining people of top 10 states according to population

latest = df.sort_values('date').groupby('States').last()
latest['unvaccinated'] = latest['population'] - latest['people_fully_vaccinated']
latest['unvaccinated'] = latest['unvaccinated']
top10 = latest.sort_values('population', ascending=False).head(10)
plt.figure(figsize=(10, 6))
plt.barh(top10.index, top10['people_fully_vaccinated'], label='Vaccinated', color='green')
plt.barh(top10.index, top10['unvaccinated'], left=top10['people_fully_vaccinated'], label='Unvaccinated', color='orange')
plt.xlabel("Number of People")
plt.title("Vaccinated vs Unvaccinated (Top 10 States)", fontsize=14, fontweight='bold')
plt.legend()
```
